```python
lst=list(data.iloc[:,48])
```

```python
def individual_count(lst,text):
    amazon=0
    flipkart=0
    paytm=0
    myntra=0
    snapdeal=0
    for i in lst:
        if "Amazon.in" in i:
            amazon+=1
        if "Flipkart.com" in i:
            flipkart+=1
        if "Paytm.com" in i:
            paytm +=1
        if "Myntra.com" in i:
            myntra +=1
        if "Snapdeal.com" in i:
            snapdeal +=1
    data = pd.DataFrame({'Company':['Amazon','Flipkart','Paytm','Myntra','SNapdeal'],'Values':[amazon,flipkart,paytm,myntra,snapdeal]})
    sns.barplot(x='Company',y='Values',data=data,order=data.sort_values('Values',ascending=False).Company)
    plt.title('Individual vote for company',fontdict={'fontname':'monospace','fontsize':20})
    plt.xlabel('Company Name ',fontdict={'fontsize':15})
    plt.ylabel('Total Votes ',fontdict={'fontsize':15})
    
    plt.style.use('default')
    plt.figure(figsize=(7,5))
    plt.pie('Values',labels='Company', data=data,explode=[.01,.01,.01,.01,.01],autopct='%1.2f%%', shadow=True)
    plt.xlabel(text,fontdict={'fontname': 'Monospace', 'fontsize': 15, 'fontweight': 'bold'})
```

```python
individual_count(lst,'Easy to use website or application choices')
```
